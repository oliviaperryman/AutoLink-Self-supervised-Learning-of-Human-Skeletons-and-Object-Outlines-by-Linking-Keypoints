```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = plt.imread('grey4/138/138_r45.png')
img_rot = plt.imread('grey4/138/138_r90.png')

w,h = img.shape
random_mask = np.random.randint(0,10,(w//8,h//8))
random_mask = cv2.resize(random_mask.astype(float), dsize=(h, w), interpolation=cv2.INTER_NEAREST)
random_mask = [[0 if x < 8 else 1 for x in row] for row in random_mask]
masked_img = img * random_mask

plt.subplot(1,3,1)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(masked_img)
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(img_rot)
plt.axis('off')
plt.show()
```

```python
# set seed 
np.random.seed(0)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 1

vmin = 0
vmax = 10

def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

for m, zlow, zhigh in [('o', 0, 10)]*10:
    xs = randrange(n, zlow, zhigh)
    ys = randrange(n, zlow, zhigh)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

# ax.axis('off')
plt.xlim(-5,15)
plt.ylim(-5,15)

# plt.show()

ax.view_init(30, 60)
# plt.show()
```
